# ðŸ“Š Funnel & CAC Analysis Notebook
This notebook analyzes leads, campaigns, and conversions data for the DT Fellowship Growth Simulation project.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
leads = pd.read_csv("../data/leads.csv")
campaigns = pd.read_csv("../data/campaigns.csv")
conversions = pd.read_csv("../data/conversions.csv")

(leads.shape, campaigns.shape, conversions.shape)
```

## ðŸ”¹ Funnel Conversion Analysis

```python

# Funnel stage distribution
funnel = leads['stage'].value_counts(normalize=True) * 100
print("Funnel conversion % by stage:")
print(funnel)

# Plot funnel as bar chart
funnel.sort_index().plot(kind='bar', figsize=(8,5), title="Funnel Conversion % by Stage")
plt.ylabel("% of Leads")
plt.show()
```

## ðŸ”¹ CAC Analysis per Channel

```python

campaigns['cost_per_conversion'] = campaigns['cost_incurred'] / campaigns['leads_generated']
campaigns[['channel', 'cost_incurred', 'leads_generated', 'cost_per_conversion']]
```

```python

campaigns.plot(x='channel', y='cost_per_conversion', kind='bar', legend=False, figsize=(8,5),
               title="Cost per Conversion by Channel")
plt.ylabel("Cost per Conversion (â‚¹)")
plt.show()
```
